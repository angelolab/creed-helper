```python
# these are one offs and dont even really need to be functions
def get_normals(normals_list, data):
    return data.loc[normals_list]
def get_tissues(normals_list):
    return list(np.unique([x.split("_")[2] for x in normals_list]))
```

```python
normals_list=["TONIC_TMA1_ln_bottom","TONIC_TMA2_ln_bottom","TONIC_TMA3_ln_bottom",
              "TONIC_TMA1_ln_top","TONIC_TMA2_ln_top","TONIC_TMA3_ln_top"]
normal_data = get_normals(normals_list,test)
tissues_list = get_tissues(normals_list)
channels = list(test.channels.values)
```

```python
for i in range(len(tissues_list)):
    samples = [x for x in normals_list if x.split("_")[2]==tissues_list[i]]
    
    for j in range(len(channels)):
        qc_metrics_per_channel = []
        
        for k in range(len(samples)):
            qc_metrics_per_channel += [qc_comp.compute_99_9_intensity(test.loc[samples[k],0:1023,0:1023,channels[j]])]
        
        ax = sns.violinplot(qc_metrics_per_channel)
        ax.set_title(tissues_list[i] + "_" + channels[j])
        ax.set(xlabel="99.9th percentile")
        plt.show()

```
