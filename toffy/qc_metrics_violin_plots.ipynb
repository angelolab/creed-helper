```python
from ark.utils.load_utils import load_imgs_from_tree, load_imgs_from_dir
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from toffy import qc_comp
from ark.utils import io_utils
import pandas as pd

def qc_metrics_plots(data_dir, 
	                 normal_tissues,
                     channels=None, 
	                 qc_metric="99.9th_percentile",
	                 fig_dir=None):
    """
    Purpose: Look for systematic bias in normal channels across TMAs for 
             each normal tissue type.
    Input: Directory containing image folders.  Normals is a string or list
            containing the normal tissue names which have to be substrings
            of the normal subfolder names.
           QC metric of choice for plotting.
    Output: Violin plots in a grid with tissue type and marker clearly labeled

    """
    # Dont load in all the tumor data, just the normals of interest.
    # may have ln_6, ln_top, ln_bottom, but these are all considered ln for
    # one plot
    normal_samples = io_utils.list_folders(dir_name=data_dir, substrs=normal_tissues)

    data = load_imgs_from_tree(data_dir=data_dir, 
	                    img_sub_folder=None, 
	                    fovs=normal_samples, 
	                    channels=None,
                        max_image_size=None)

    if not channels:
        channels = list(data.channels.values)

    # need to make a nice pandas DF with columns: sample, channel, 99.9th value
    for i in range(len(normal_tissues)):

        # For each tissue type, get all the samples
        samples = [x for x in normal_samples if normal_tissues[i] in x]
        plotting_df = pd.DataFrame(columns=["sample",'channel',"99.9th_percentile"])
        

        for j in range(len(channels)):
            # Channel is a gene
            qc_metrics_per_channel = []

            for k in range(len(samples)):
                # go through all samples per gene. list should be of length==samples
                qc_metrics_per_channel += [[normal_tissues[i],
                                           channels[j],
                                           qc_comp.compute_99_9_intensity(data.loc[samples[k], 0:1023, 0:1023, channels[j]])]]
            

            plotting_df = plotting_df.append( pd.DataFrame(qc_metrics_per_channel,columns=plotting_df.columns) )

        # plt.show()
        plt.figure(figsize=(16,3))
        ax = sns.violinplot(x="channel",y="99.9th_percentile",data=plotting_df,inner=None,scale="width")
        ax = sns.swarmplot(x="channel",y="99.9th_percentile",data=plotting_df,color="black",edgecolor="gray")
        ax.set_title(normal_tissues[i])
        ax.set_xticklabels([x[0:4] for x in channels])


        # Probaly dont need plt.show() if its a jupe notebook
        if fig_dir:
            plt.savefig(fig_dir+normal_tissues[i]+"_qc.png",dpi=300)
        plt.show()
        plt.close()


    return


qc_metrics_plots(data_dir="/Users/heskett/control_sample_Examples",
	             normal_tissues=["ln"],
                 channels=["CD3","CD4","CD8","CD11c","CD14",
                           "CD20","CD31","CD38","CD45","CD45RB","CD45RO",
                           "CD56","CD57","CD68","CD69","CD163","Calprotectin","ECAD",
                            "CK17","HLA1","Fibronectin","FOXP3","PDL1","TCF1","SMA","TBET",
                            "TIM3","Vim","LAG3","IDO","HLADR","H3K9ac"],
                fig_dir="/Users/heskett/toffy/")



```
