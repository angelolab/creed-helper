### Inspecting Median Pulse Height Data
This notebook can be used to manually inspect MPH data to verify everything is performing as expected.

```python
import sys 
sys.path.append('../') 
```

```python
import os
import pandas as pd

from toffy import mph_inspect
```

### Define file parameters

- `run_name` should contain the exact name of the MIBI run to locate the mph data from

```python
# the name of the run
run_name = 'YYYY-MM-DD_run_name'

# path where plots will get saved
bin_file_dir = os.path.join('D:\\Data', run_name)
```

### Inputs for computing MPH intensities

- `mass` given the mass for the panel
- `mass_start` is the beginning of mass integration range
- `mass_stop` is the end of mass integration range
- `fov_list` details exactly which FOVs you would like to generate histograms for, 
    if left as None, will generate histograms for all FOV bin files in the `bin_file_dir`

```python
# input mass specs
mass = 98
mass_start=97.5
mass_stop=98.5

# optional argument
fov_list = None
```

### Visualize MPH intensity histograms

```python
# plot data
mph_inspect.visualize_mph_hist(bin_file_dir, mass,mass_start, mass_stop, 
                               fov_list, normalize=False)
```

```python
# plot normalized data
mph_inspect.visualize_mph_hist(bin_file_dir, mass, mass_start, mass_stop, 
                               fov_list, normalize=True)
```
