### Inspecting Pulse Height Data
This notebook can be used to manually inspect the pulse height intensities to verify everything is performing as expected.

```python
import os
from toffy import mph_inspect
```

### Define file parameters

- `run_name`: should contain the exact name of the MIBI run to locate the intensity data from
- `bin_file_path`: the directory containing your bin files

```python
# set up directories for the run
run_name = 'YYYY-MM-DD_run_name'

bin_file_dir = os.path.join('D:\\Data', run_name)
```

### Inputs for computing intensities

- `mass`: the mass for the panel
- `mass_start`: beginning of mass integration range
- `mass_stop`: end of mass integration range
- `fov_list`: details exactly which FOVs to generate histograms for, if left as None, will generate histograms for all FOV bin files in the `bin_file_dir`

```python
# input mass specs
mass = 98
mass_start=97.5
mass_stop=98.5

# optional argument
fov_list = None

# fov_list = []
```

### Visualize intensity histograms

```python
# plot data
mph_inspect.visualize_intensity_data(bin_file_dir, mass,mass_start, mass_stop, fov_list, normalize=False)
```

```python
# plot normalized data
mph_inspect.visualize_intensity_data(bin_file_dir, mass, mass_start, mass_stop, fov_list, normalize=True)
```
